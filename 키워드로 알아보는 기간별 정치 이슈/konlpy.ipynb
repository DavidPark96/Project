```python
import re
def mecab_word(text):
    useless = ['VV','JKS','JKG','JKO','JKB','JKV','JKQ','JC','JX','ETN','VCP+EC','XSA+ETM','XSA+EC+VCP','VCP','EC','ETM','VCP+EP','VA','XSV','MAG','EC','XSA+EC','EP','VCP+ETM','XSV+EC','VX+EP','XSV+EP','VX+EC','JKS+ETN+JX','VCP+ETN']
    text = re.sub(r'[()\'"‘’“?!.,]',' ',text)
    split_text = text.split() #텍스트 띄어쓰기 단위로 나누기
    for word in split_text: #단어들을 하나씩 처리
        pos = mecab.pos(word) #단어 하나씩 형태소 분리 : 이 행위의 이점은 명사, 명사+다른 형태소의 모양을 구분해낼 수 있음.
        for i in pos[::-1]: #역순으로 하나씩 확인해서 단어의 맨 뒤가 명사로 분리되지 않은 불용어들 순차적으로 제거
            if i[1] in useless:
                pos.pop(-1)
            else:
                break

        if len(pos) >= 1:
            if pos[-1][1] not in useless: #띄어쓰기 단위별 단어의 끝에 조사(JK) 혹은 불용어들이 많아 이러한 단어는 제거. (NNG+JK 등)
    #             print([i[1] for i in pos])
                for j in (i[1] for i in pos): #남은 단어 조합 중에 명사가 없는 단어 색출.
                    if j in ['NNP','NNG','NNB','NNBC']: #명사 형태소
                        data = "".join((i[0] for i in pos))  #남은 단어들은 형태소를 조합
                        if len(data) > 1: #한글자 단어 삭제
                            print(data)
                        break #단어의 형태소를 순회하기 때문에 break를 걸지 않으면 형태소 개수만큼 for문이 돌아서 단어가 증가한다.
```

```python
def sentence2word(sentence):
    words = han.pos(sentence)
    
    data = []
    for word in words:
        if word[1] in ['N', 'M'] and len(word[0]) >= 2:
            data.append(word[0])
        elif word[1] in ['P']:
            okt_word = okt.pos(word[0],stem=True)[0]
            if okt_word[1] == 'Verb':
                data.append(okt_word[0])
                
    result = Counter(data)
    return [(key, result[key]) for key in result]
```

# 1. konlpy 클래스 비교 (Okt, Mecab, Hannanum)

## 비교 함수 선언

```python
def comparsion_konlpy(sentence):
    t= time.time()
    df1 = pd.DataFrame(mecab.pos(sentence),columns=['Mecab','형태소'])
    print(f'Mecab : {time.time() - t:.5f}')
    
    t= time.time()
    df2 = pd.DataFrame(okt.pos(sentence,stem=True),columns=['Okt','형태소'])
    print(f'Okt : {time.time() - t:.5f}')
    
    t= time.time()
    df3 = pd.DataFrame(han.pos(sentence),columns=['Hannanum','형태소'])
    print(f'Hannanum : {time.time() - t:.5f}')
    return pd.concat([df1,df2,df3], axis=1)
```
